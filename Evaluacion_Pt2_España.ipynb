```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv
españa= pd.read_csv("espana_limpio.csv", encoding = 'unicode_escape')
españa.head(5)
```

**Obtener los registros que cumplan las siguientes condiciones para cada dataframe:**

```python
# a) host_acceptance_rate> 50%
filtro1_esp = españa.loc[españa['host_acceptance_rate'] > 0.50]
filtro1_esp
```

```python
# b) Sólo los registros con categoría “superhost”
filtro2_esp = españa.loc[españa['host_is_superhost'] == "t"]
filtro2_esp
```

```python
# c) Los registros que no hallan verificado identidad “ not identity_verified”
filtro3_esp = españa.loc[españa['host_identity_verified'] == "f"]
filtro3_esp
```

```python
# d) Los registro cuyo property_type = “Private room” y “Hotel room”
filtro4_esp = españa[~(españa["property_type"] == "Private room") & ~(españa["property_type"] == "Hotel room")]
filtro4_esp
```

```python
# e) Los registros que cuenten con bathroom > 1
filtro5_esp = españa.loc[españa['bathrooms'] > 1]
filtro5_esp
```

```python
# f) Los registros cuyo precio sea mayor de $10,000 y que sean de tipo “Entire home”
filtro6_esp = españa[~(españa["price"] > 10000) & ~(españa["property_type"] == "Entire home")]
filtro6_esp
```

```python
# g) Los registros cuyo review_scores_cleanliness > 4.5
filtro7_esp = españa.loc[españa['review_scores_cleanliness'] > 4.5]
filtro7_esp
```

```python
# h) Los registros cuyo review_scores_value > 4.9
filtro8_esp = españa.loc[españa['review_scores_value'] > 4.9]
filtro8_esp
```

```python
# i) Los registros cuya availability_365< 100
filtro9_esp = españa.loc[españa['availability_365'] < 100]
filtro9_esp
```

```python
# j) Los registros cuya host_response_time sea “within an hour”
filtro10_esp = españa.loc[españa['host_response_time'] == "within an hour"]
filtro10_esp
```

```python
# Convertir a csv cada filtro
filtro1_esp.to_csv("filtro1_esp.csv")
filtro2_esp.to_csv("filtro2_esp.csv")
filtro3_esp.to_csv("filtro3_esp.csv")
filtro4_esp.to_csv("filtro4_esp.csv")
filtro5_esp.to_csv("filtro5_esp.csv")
filtro6_esp.to_csv("filtro6_esp.csv")
filtro7_esp.to_csv("filtro7_esp.csv")
filtro8_esp.to_csv("filtro8_esp.csv")
filtro9_esp.to_csv("filtro9_esp.csv")
filtro10_esp.to_csv("filtro10_esp.csv")
```
