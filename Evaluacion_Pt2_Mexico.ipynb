```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv
mexico = pd.read_csv("Mexico_limpio.csv", encoding = 'unicode_escape')
mexico.head(5)
```

**Obtener los registros que cumplan las siguientes condiciones para cada dataframe:**

```python
# a) host_acceptance_rate> 50%
filtro1_mex = mexico.loc[mexico['host_acceptance_rate'] > 0.50]
filtro1_mex
```

```python
# b) Sólo los registros con categoría “superhost”
filtro2_mex = mexico.loc[mexico['host_is_superhost'] == "t"]
filtro2_mex
```

```python
# c) Los registros que no hallan verificado identidad “ not identity_verified”
filtro3_mex = mexico.loc[mexico['host_identity_verified'] == "f"]
filtro3_mex
```

```python
# d) Los registro cuyo property_type = “Private room” y “Hotel room”
filtro4_mex = mexico[~(mexico["property_type"] == "Private room") & ~(mexico["property_type"] == "Hotel room")]
filtro4_mex
```

```python
# e) Los registros que cuenten con bathroom > 1
filtro5_mex = mexico.loc[mexico['bathrooms'] > 1]
filtro5_mex
```

```python
# f) Los registros cuyo precio sea mayor de $10,000 y que sean de tipo “Entire home”
filtro6_mex = mexico[~(mexico["price"] > 10000) & ~(mexico["property_type"] == "Entire home")]
filtro6_mex
```

```python
# g) Los registros cuyo review_scores_cleanliness > 4.5
filtro7_mex = mexico.loc[mexico['review_scores_cleanliness'] > 4.5]
filtro7_mex
```

```python
# h) Los registros cuyo review_scores_value > 4.9
filtro8_mex = mexico.loc[mexico['review_scores_value'] > 4.9]
filtro8_mex
```

```python
# i) Los registros cuya availability_365< 100
filtro9_mex = mexico.loc[mexico['availability_365'] < 100]
filtro9_mex
```

```python
# j) Los registros cuya host_response_time sea “within an hour”
filtro10_mex = mexico.loc[mexico['host_response_time'] == "within an hour"]
filtro10_mex
```

```python
# Convertir a csv cada filtro
filtro1_mex.to_csv("filtro1_mex.csv")
filtro2_mex.to_csv("filtro2_mex.csv")
filtro3_mex.to_csv("filtro3_mex.csv")
filtro4_mex.to_csv("filtro4_mex.csv")
filtro5_mex.to_csv("filtro5_mex.csv")
filtro6_mex.to_csv("filtro6_mex.csv")
filtro7_mex.to_csv("filtro7_mex.csv")
filtro8_mex.to_csv("filtro8_mex.csv")
filtro9_mex.to_csv("filtro9_mex.csv")
filtro10_mex.to_csv("filtro10_mex.csv")
```
