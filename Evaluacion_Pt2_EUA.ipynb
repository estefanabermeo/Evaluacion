```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv
eua= pd.read_csv("eua_limpio.csv", encoding = 'unicode_escape')
eua.head(5)
```

**Obtener los registros que cumplan las siguientes condiciones para cada dataframe:**

```python
# a) host_acceptance_rate> 50%
filtro1_eua = eua.loc[eua['host_acceptance_rate'] > 0.50]
filtro1_eua
```

```python
# b) Sólo los registros con categoría “superhost”
filtro2_eua = eua.loc[eua['host_is_superhost'] == "t"]
filtro2_eua
```

```python
# c) Los registros que no hallan verificado identidad “ not identity_verified”
filtro3_eua = eua.loc[eua['host_identity_verified'] == "f"]
filtro3_eua
```

```python
# d) Los registro cuyo property_type = “Private room” y “Hotel room”
filtro4_eua = eua[~(eua["property_type"] == "Private room") & ~(eua["property_type"] == "Hotel room")]
filtro4_eua
```

```python
# e) Los registros que cuenten con bathroom > 1
filtro5_eua = eua.loc[eua['bathrooms'] > 1]
filtro5_eua
```

```python
# f) Los registros cuyo precio sea mayor de $10,000 y que sean de tipo “Entire home”
filtro6_eua = eua[~(eua["price"] > 10000) & ~(eua["property_type"] == "Entire home")]
filtro6_eua
```

```python
# g) Los registros cuyo review_scores_cleanliness > 4.5
filtro7_eua = eua.loc[eua['review_scores_cleanliness'] > 4.5]
filtro7_eua
```

```python
# h) Los registros cuyo review_scores_value > 4.9
filtro8_eua = eua.loc[eua['review_scores_value'] > 4.9]
filtro8_eua
```

```python
# i) Los registros cuya availability_365< 100
filtro9_eua = eua.loc[eua['availability_365'] < 100]
filtro9_eua
```

```python
# j) Los registros cuya host_response_time sea “within an hour”
filtro10_eua = eua.loc[eua['host_response_time'] == "within an hour"]
filtro10_eua
```

```python
# Convertir a csv cada filtro
filtro1_eua.to_csv("filtro1_eua.csv")
filtro2_eua.to_csv("filtro2_eua.csv")
filtro3_eua.to_csv("filtro3_eua.csv")
filtro4_eua.to_csv("filtro4_eua.csv")
filtro5_eua.to_csv("filtro5_eua.csv")
filtro6_eua.to_csv("filtro6_eua.csv")
filtro7_eua.to_csv("filtro7_eua.csv")
filtro8_eua.to_csv("filtro8_eua.csv")
filtro9_eua.to_csv("filtro9_eua.csv")
filtro10_eua.to_csv("filtro10_eua.csv")
```
