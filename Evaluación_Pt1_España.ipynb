### **Reemplazo de valores nulos**

```python
#Elimino columnas que tienen todas las filas vacías
españa.drop(["neighbourhood_group_cleansed", "bathrooms", "calendar_updated"], axis=1)
```

Para las columnas con **datos cualitativos** reemplazaré los valores nulos por strings en concretos que sean coherentes al contexto del atributo.

```python
españa["description"] = españa["description"].fillna("Sin descripción") 
españa["neiborhood_overview"] = españa["neighborhood_overview"].fillna("Sin información")
españa["host_name"] = españa["host_name"].fillna("Anónimo")
españa["host_since"] = españa["host_since"].fillna("Sin fecha")
españa["host_location"] = españa["host_location"].fillna("Sin ubicación")
españa["host_about"] = españa["host_about"].fillna("Sin información")
españa["host_response_time"] = españa["host_response_time"].fillna("Sin información")
españa["host_is_superhost"] = españa["host_is_superhost"].fillna("Sin información")
españa["host_thumbnail_url"] = españa["host_thumbnail_url"].fillna("Sin url")
españa["host_picture_url"] = españa["host_picture_url"].fillna("Sin url")
españa["host_thumbnail_url"] = españa["host_thumbnail_url"].fillna("Sin url")
españa["host_has_profile_pic"] = españa["host_has_profile_pic"].fillna("Sin información")
españa["host_identity_verified"] = españa["host_identity_verified"].fillna("Sin información")
españa["neighbourhood"] = españa["neighbourhood"].fillna("Sin ubicación")
españa["bathrooms_text"] = españa["bathrooms_text"].fillna("Sin información")
españa["first_review"] = españa["first_review"].fillna("Sin fecha")
españa["last_review"] = españa["last_review"].fillna("Sin fecha")
españa["license"] = españa["license"].fillna("Sin licencia")
españa["host_response_rate"] = españa["host_response_rate"].fillna("0%")
españa["host_acceptance_rate"] = españa["host_acceptance_rate"].fillna("0%")
```

Para las columnas con **datos cuantitativos** remplazaré los valores nulos por 0 para no alterar los resultados y el análisis de este.

```python
españa = españa.fillna(0)
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=españa_num
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Outliers del Dataframe
outliers= españa_num[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
españa2 = españa_num[(y<=Limite_Superior)&(y>=Limite_Inferior)]
españa2
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=españa_num.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=españa_num.copy()
data_clean=data_clean.fillna(round(españa_num.mean(),1))
data_clean
```
